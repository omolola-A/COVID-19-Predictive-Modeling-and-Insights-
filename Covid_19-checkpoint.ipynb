# Trends Over Time

```python
print(df['Year'].unique())  # This will print the unique years in your dataset
```

```python
# Ensuring 'Date' is in datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Filter data for the year 2020
df_2020 = df[df['Date'].dt.year == 2020]

# Group by month and aggregate metrics
df_2020['Month'] = df_2020['Date'].dt.strftime('%B')  # Extract month names
df_monthly_2020 = df_2020.groupby('Month').agg({'Confirmed': 'sum', 'Deaths': 'sum', 'Recovered': 'sum'}).reset_index()

# To maintain the chronological order of months
months_order = ['January', 'February', 'March', 'April', 'May', 'June', 
                'July', 'August', 'September', 'October', 'November', 'December']
df_monthly_2020['Month'] = pd.Categorical(df_monthly_2020['Month'], categories=months_order, ordered=True)
df_monthly_2020.sort_values('Month', inplace=True)
```

```python
# Trends for confirmed, deaths, and recovered cases over months in 2020
plt.figure(figsize=(12, 6))
plt.plot(df_monthly_2020['Month'], df_monthly_2020['Confirmed'], label='Confirmed Cases', marker='o')
plt.plot(df_monthly_2020['Month'], df_monthly_2020['Deaths'], label='Deaths', marker='o')
plt.plot(df_monthly_2020['Month'], df_monthly_2020['Recovered'], label='Recovered', marker='o')

plt.xlabel('Month')
plt.ylabel('Number of Cases')
plt.title('COVID-19 Monthly Trends in 2020')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### The trends graph shows confirmed, death, and recovery cases over seven months. No cases were recorded from January to March. From March to July, an increase in confirmed cases corresponded with a rise in both deaths and recoveries. Notably, deaths exceeded recoveries during this period, but by the end of July, recoveries surpassed the death rate.

# Time-Series Modeling: ARIMA Model

```python
df = df.sort_values(by='Date')
```

```python
time_series_data = df.groupby('Date')['Confirmed'].sum().reset_index()
```

```python
time_series_data['Date'] = pd.to_datetime(time_series_data['Date'])
time_series_data.set_index('Date', inplace=True)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(time_series_data['Confirmed'], label='Confirmed Cases', color='blue')
plt.title('COVID-19 Confirmed Cases Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Cases')
plt.legend()
plt.show()
```

## Time Series Decomposition

```python
from statsmodels.tsa.seasonal import seasonal_decompose

result = seasonal_decompose(time_series_data['Confirmed'], model='additive')
result.plot()
plt.show()
```

## Checking for Stationarity

```python
from statsmodels.tsa.stattools import adfuller

adf_test = adfuller(time_series_data['Confirmed'])
print(f"ADF Statistic: {adf_test[0]}")
print(f"p-value: {adf_test[1]}")
```

```python
time_series_data['Confirmed_diff'] = time_series_data['Confirmed'].diff().dropna()
```

## Dataset Splitting

```python
train = time_series_data.iloc[:-30]  # Use all except the last 30 days for training
test = time_series_data.iloc[-30:]  # Use the last 30 days for testing
```

```python
# !pip install pmdarima
```

## Build the ARIMA ModelBuild the ARIMA Model

```python
from pmdarima import auto_arima

model = auto_arima(train['Confirmed'], seasonal=False, trace=True)
print(model.summary())
```

### Determining the p, d, and q Parameters

```python
from pmdarima import auto_arima

auto_model = auto_arima(train['Confirmed'], seasonal=False, trace=True)
print(auto_model.summary())
```

```python
from statsmodels.tsa.arima.model import ARIMA

arima_model = ARIMA(train['Confirmed'], order=(2, 2, 1))
arima_result = arima_model.fit()
print(arima_result.summary())
```

## Forecasting Future Values

```python
predictions = arima_result.forecast(steps=len(test))
```

```python
plt.figure(figsize=(12, 6))
plt.plot(train['Confirmed'], label='Train', color='blue')
plt.plot(test['Confirmed'], label='Test', color='orange')
plt.plot(test.index, predictions, label='Predictions', color='green')
plt.legend()
plt.show()
```

## Accuracy Metrics

```python
from sklearn.metrics import mean_squared_error

rmse = mean_squared_error(test['Confirmed'], predictions, squared=False)
print(f"RMSE: {rmse}")
```

### The Root Mean Squared Error (RMSE) of the model is 0.1306, indicating that, on average, the model's predicted values deviate from the actual values by approximately 0.13 units. This low RMSE value suggests that the ARIMA model performs well in forecasting the confirmed COVID-19 cases, with minimal error in its predictions.

```python
future_forecast = arima_result.forecast(steps=30)  # Predict the next 30 days
print(future_forecast)
```

```python
import matplotlib.pyplot as plt

# Create a date range for the forecast
forecast_index = pd.date_range(start=test.index[-1] + pd.Timedelta(days=1), periods=30, freq='D')

# Plot historical data and the forecast
plt.figure(figsize=(12, 6))
plt.plot(train.index, train['Confirmed'], label='Training Data')
plt.plot(test.index, test['Confirmed'], label='Test Data')
plt.plot(forecast_index, future_forecast, label='Forecast (Next 30 Days)', color='red')
plt.xlabel('Date')
plt.ylabel('Confirmed Cases')
plt.title('COVID-19 Forecast')
plt.legend()
plt.grid()
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# Calculate metrics
mae = mean_absolute_error(test['Confirmed'], arima_result.forecast(steps=len(test)))
rmse = np.sqrt(mean_squared_error(test['Confirmed'], arima_result.forecast(steps=len(test))))

print(f"Mean Absolute Error (MAE): {mae}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
```

### The Mean Absolute Error (MAE) of 0.0971 indicates that, on average, the model's predictions deviate from the actual values by approximately 0.10 units. The Root Mean Squared Error (RMSE) of 0.1306 further highlights the model's accuracy, showing a low level of error in its forecasts. Together, these metrics suggest that the ARIMA model provides reliable predictions for confirmed COVID-19 cases with minimal deviations.

```python
def calculate_mape(actual, forecast):
    mape = np.mean(np.abs((actual - forecast) / actual)) * 100
    return mape

# Calculate MAPE for your model's predictions
mape = calculate_mape(test['Confirmed'], arima_result.forecast(steps=len(test)))
print(f"Mean Absolute Percentage Error (MAPE): {mape}%")
```

### The Mean Absolute Percentage Error (MAPE) of 2.84% indicates that the model's predictions are, on average, within 2.84% of the actual values. This low percentage error demonstrates that the ARIMA model provides highly accurate forecasts for confirmed COVID-19 cases.

```python
from sklearn.metrics import r2_score

# Calculating R² for model's predictions
r2 = r2_score(test['Confirmed'], arima_result.forecast(steps=len(test)))
print(f"R-squared (R²): {r2}")
```

### The R-squared value of 0.9142 indicates that approximately 91.42% of the variance in the actual confirmed COVID-19 cases is explained by the ARIMA model. This high R² value suggests that the model fits the data well and has strong predictive power.
